```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import kurtosis
from scipy.stats import skew
from scipy.stats import norm

import matplotlib.pyplot as plt
%matplotlib inline  
import seaborn as sns
```

**Загружаем таблицу**

```python
wines=pd.read_csv('wine_data.csv')
wines.head(3)
```

```python
wines.info()
```

```python
wines.describe()
```

**Получаем таблицу корреляций**

```python
wines_corr=wines.corr()
wines_corr
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(wines_corr, ax=ax, annot=True, cmap="YlGnBu", linewidths=.5,
            xticklabels=wines_corr.columns.values,
            yticklabels=wines_corr.columns.values)
```

```python
sns.pairplot(wines, vars=['Alcohol', 'Proline', 'Flavanoids', 'Total phenols'])
```

**t-test**

```python
stats.ttest_1samp(wines['Alcohol'], 13)
```

```python
alcohol_1 = wines[wines['Class'] == 1]['Alcohol']
alcohol_2 = wines[wines['Class'] == 3]['Alcohol']

stats.ttest_ind(alcohol_1, alcohol_2)
```

```python
wines.groupby('Class').mean()
```

```python

stats.ttest_ind(wines['Proline'], wines['Alcohol'])
```
